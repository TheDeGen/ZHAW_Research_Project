# Exploratory Data Analysis of News Headlines

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

## Load the datasets

```python
cnbc_df = pd.read_csv('../data/raw/cnbc_headlines.csv')
guardian_df = pd.read_csv('../data/raw/guardian_headlines.csv')
reuters_df = pd.read_csv('../data/raw/reuters_headlines.csv')

cnbc_df['source'] = 'CNBC'
guardian_df['source'] = 'The Guardian'
reuters_df['source'] = 'Reuters'

df = pd.concat([cnbc_df, guardian_df, reuters_df], ignore_index=True)
```

## Basic Data Analysis

```python
print(f"Total number of headlines: {len(df)}")
print("Number of headlines from each source:")
print(df['source'].value_counts())
```

```python
print("Missing values:")
print(df.isnull().sum())
```

## Visualize the number of headlines from each source

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='source', data=df, palette='viridis')
plt.title('Number of Headlines from Each Source')
plt.xlabel('Source')
plt.ylabel('Number of Headlines')
plt.show()
```
